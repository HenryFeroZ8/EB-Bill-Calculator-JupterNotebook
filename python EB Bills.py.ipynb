```python
def EB(name):
    year = int(input("Enter Year"))
    m = {i:j for i,j in zip(range(1,13),['jan','feb','mar','apr','may','jun','july','aug','sep','oct','nov','dec'])}
    for i,j in m.items():
        print(i,'=',j)

    
    month = m[int(input('enter month number'))]
    units = float(input("Enter Totel units consumed"))
    if units<100:
        amt = 0
        sur = 0
    elif 100<units<=200:
        amt = 0 + (units-100)*2
        sur = 10
    elif 200<units<=300:
        amt = 0 + 200 + (units-200)*3
        sur = 20
    elif 300<units<=400:
        amt = 0 + 200 + 300 +(units-300)*4
        sur = 30
    else:
        amt = 0 + 200 + 300 + 400 + (units-400)*5
        sur = 50
        
    print(f'Hello user ,Your totel bill for {units} units is rs.{amt+sur} ')
    if year in data[name]['EB_bills'].keys():
        
        data[name]['EB_bills'][year][month]={}
        data[name]['EB_bills'][year][month]['units']=units
        data[name]['EB_bills'][year][month]['bill']=amt+sur
    else:
         data[name]['EB_bills'][year]={}
         data[name]['EB_bills'][year][month]={}
         data[name]['EB_bills'][year][month]['units']=units
         data[name]['EB_bills'][year][month]['bill']=amt+sur
    options(name)
    
```
